<h4 style="text-align: center;">Part 2. Missing Data</h4>
<p style="text-align: center;">This dataframe initially contains 1994 rows. First, we drop the columns that are not predictive: state, county, community, and fold. We leave in communityname for now to make this easier to map, though it will not be used for prediction. If we drop every column with more than 200  ?'s, which represents missing data from the dataframe, we end up with 102 columns. We will now impute the data, which means we would add a represented/typical value where missing points occur. </p>

Next, we want to iterate through each column, besides the communityName since its the only attribute that is a string (we will leave it in our dataframe for now). During each iteration, we will convert each cell in the dataframe that contains a missing value ("?") with NumPy's NaN object. This represents an object that is 'Not A Number'. We can easily drop the columns that have 200 or more missing data entries using Pandas' to_numeric function with the 'coerce' parameter, which will convert all the ?'s to NaN's. 

```python
for col in data.drop('communityname',1):
    data[col]= pd.to_numeric(data[col], errors='coerce')
    if(data[col].isnull().sum() >= 200):
        data.drop(col, 1, inplace=True)
```

Now that we have dropped sparse columns, we will want to impute the rest of the missing data. To do this, we use the current attributes mean and populate the missing data with that mean. This is the mathematically closest operation to do to find the expected value's of the missing entries. We can make use of Pandas' fillna function, which will take in a Series (a column of our data) and fill in all the NaN's with the mean of that column. The fillna function requires that missing data be in the format of NA/NaN values, so that is why we went through the "trouble" to convert missing entries from '?' to NaN objects. 

```python
for col in data.drop('communityname',1):
    mean = data[col].mean()
    data[col].fillna(value=mean,inplace=True)
data.head()
```

Note how originial dataframe had 128 columns, now we only have 102 columns. Now our data is tidy, we can do some analysis, visualization and machine learning to predict crime.

<h4 style="text-align: center;">Regression</h4>
<p style="text-align: center;">We choose to use two types of linear regression and two types of non-linear to see if the data was linear. The regressions we did were: Linear Regression,  Stochastic Gradient Descent Regression, Extra Trees, and Random Forests. Linear regression uses ordinary least squares to fit a line to the data. Stochastic gradient descent mimimizes squared loss while adding a regularizing penalty function to promote sparsity in the weight vector. Specifically, we used the elastic net penalty function, which is a combination of an L1 and L2 loss function. For further reading on how this function works, please see <a href="http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDRegressor.html"> this link</a>.
Extra Trees and Random Forests combine a set of randomized decision trees on various subsets of the data. They average these models to help with over-fitting and improve accuracy.
Random Forests differ from Extra Trees in how they choose to split and sample the data given. Extra Trees tend to have more leaves, while Random Forests allow for sampling with replacement.

```python
violentCrimesPerPop = data['ViolentCrimesPerPop']
violentCrimesPerPop = violentCrimesPerPop.as_matrix().flatten() #y-value
violentCrimesSave = violentCrimesPerPop
#Create 
data_df = data
data_arr = data.drop(['communityname','ViolentCrimesPerPop'],1).as_matrix().flatten().reshape(1994,100)
data_arr = data_arr.astype(np.float)
```

It is good practice to use 80% of your data for training your model and 20% to test your model. The below code splits our data into training data and testing data. This is how we will train and explore our machine learning algorithms.

```python
rand_permute = np.random.permutation(len(data_arr))
data_arr = data_arr[rand_permute,:]
violentCrimesPerPop = violentCrimesPerPop[rand_permute]

training_size = int(np.floor(len(data_arr)*.8))
test_size = len(data_arr) - training_size

training_data_arr = data_arr[0:training_size,:]
training_violent_crimes = violentCrimesPerPop[0:training_size]

test_data_arr = data_arr[training_size+1:-1,:]
test_violent_crimes = violentCrimesPerPop[training_size+1:-1]
```

We will use linear regression, stochastic gradient descent regression, extra trees regression, and random forests for our machine learning process. We also sort the feature importances that we calculated using the extra trees and random forest regressor's.

```python
from sklearn import linear_model
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import RandomForestRegressor
from IPython.core import display as ICD
from IPython.display import display, HTML

reg = linear_model.LinearRegression()
reg.fit(training_data_arr, training_violent_crimes)
linear_model.LinearRegression(copy_X=True, fit_intercept=True, n_jobs=1, normalize=False)

#Stochastic Gradient Descent
SGD = SGDRegressor(penalty='elasticnet')
SGD = SGD.fit(training_data_arr, training_violent_crimes)

#Extra Tree Regressor
ETReg = ExtraTreesRegressor()
ETReg.fit(training_data_arr, training_violent_crimes)

#Find most valuable feature for predicting crime rates
a = ETReg.feature_importances_
sorted_indET = np.argsort(a)

#Random Forest Regressor
RFReg = RandomForestRegressor()
RFReg.fit(training_data_arr, training_violent_crimes)

#Find most valuable feature for predicting crime rates
b = RFReg.feature_importances_
sorted_indRF = np.argsort(b)

_datacrimePredictRFReg = RFReg.predict(test_data_arr)
_datacrimePredictETReg = ETReg.predict(test_data_arr)
_datacrimePredictSGD = SGD.predict(test_data_arr)
_datacrimePredictLinReg = reg.predict(test_data_arr)
```
